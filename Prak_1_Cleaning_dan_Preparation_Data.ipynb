```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
netflix = pd.read_csv('netflix_titles.csv')
netflix
```

```python
netflix.shape
```

```python
netflix.dtypes
```

```python
netflix.isnull().sum()
```

```python
sns.heatmap(netflix.isnull(), cbar=False)
```

```python
netflix.drop('director',axis=1).head()
```

```python
#drop row

no_director=netflix[netflix['director'].isnull()].index
netflix.drop(no_director, axis=0)
```

```python
netflix[netflix['rating'].isnull()]
```

```python
mode=''.join(netflix['rating'].mode())
netflix['rating'].fillna(mode, inplace=True)
mode
```

```python
netflix['duration'].fillna(0, inplace=True)
```

```python
df_movie =netflix[netflix['type']=='Movie']
df_movie
```

```python
df_movie['duration']
```

```python
df_movie= df_movie.assign(minute= df_movie['duration'].str.extract(r'(\d+)',expand=False).dropna().astype(int))
df_movie
```

```python
fig, ax=plt.subplots(nrows=1, ncols=1)
ax = sns.boxplot(x=df_movie['minute'])
fig.tight_layout()
```

```python
#outliers df_movie[df_movie['minute']<50 | df_movie['minute']>150]
#filtering outliers out
df_movie=df_movie[(df_movie['minute']>50) & (df_movie['minute']<150)]
```
